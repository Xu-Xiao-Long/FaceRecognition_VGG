```python
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import cv2
```

```python
def cropped_image(imagePath,savePath):
    face_detector = cv2.CascadeClassifier('D:/opencv/opencv/sources/data/haarcascades/haarcascade_frontalface_alt2.xml')

    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(gray,1.12,5)

    print("Found {0} faces!".format(len(faces)))
    
    if len(faces)!= 0:
        for x,y,w,h in faces:
            cv2.rectangle(image,(x,y),(x+w,y+h),color=(255,0,0),thickness=2)
        
        im = Image.open(imagePath)
        
        if im.mode in ("RGBA", "P"):
            im = im.convert("RGB")

        (x, y, w, h) = faces[0]
        center_x = x+w/2
        center_y = y+h/2
        b_dim = min(max(w,h)*1.2,im.width, im.height) # WARNING : this formula in incorrect
        #box = (x, y, x+w, y+h)
        box = (center_x-b_dim/2, center_y-b_dim/2, center_x+b_dim/2, center_y+b_dim/2)
        # Crop Image
        crpim = im.crop(box).resize((224,224))
        plt.imshow(np.asarray(crpim))
        crpim.save(savePath)
    else:
        return

```

```python
import os
import tensorflow as tf
from keras.models import Model
from keras.layers import Dense
from keras import optimizers
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
trdata = ImageDataGenerator()
traindata = trdata.flow_from_directory(directory="D:/AXXL/FaceData/star_face/train", target_size=(224,224))

tsdata = ImageDataGenerator()
testdata = tsdata.flow_from_directory(directory="D:/AXXL/FaceData/star_face/test", target_size=(224,224))
```

```python
from keras.applications.vgg16 import VGG16
vggmodel = VGG16(weights='imagenet', include_top=True)
```

```python
for layers in (vggmodel.layers)[:19]:
    print(layers)
    layers.trainable = False
```

```python
X = vggmodel.layers[-2].output
predictions = Dense(6, activation="softmax")(X)
model_final = Model(inputs = vggmodel.input, outputs = predictions)
```

```python
import tensorflow as tf
model_final.compile(
    loss = "categorical_crossentropy", 
    optimizer = tf.keras.optimizers.SGD(lr=0.0001, momentum=0.9), 
    metrics=["accuracy"])
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping

checkpoint = ModelCheckpoint("vgg16_star_face.h5", 
                             monitor='val_accuracy',
                            verbose=1,
                             save_best_only=True,
                             save_weights_only=False,
                             mode='auto',
                             period=1
                            )
early = EarlyStopping(monitor='val_accuracy',
                     min_delta=0,
                     patience=40,
                     verbose=1,
                     mode='auto')

hist = model_final.fit_generator(generator = traindata,
                         steps_per_epoch = 2,
                         epochs = 100,
                         validation_data = testdata,
                         validation_steps = 1,
                         callbacks = [checkpoint,early])
```

```python
import matplotlib.pyplot as plt
plt.plot(hist.history["accuracy"])
plt.plot(hist.history['val_accuracy'])
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title("model accuracy")
plt.ylabel("Accuracy")
plt.xlabel("Epoch")
plt.legend(["Accuracy","Validation Accuracy","loss","Validation Loss"])
plt.show()
```
